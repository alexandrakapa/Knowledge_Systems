````python
def get_place_by_imdb_id(imdb_id): #a function to return narrative placelabel based on imdb id
  imdb_id = '"'+imdb_id+'"'
  endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"
  query = """ SELECT ?placeLabel
    WHERE {
        ?movie wdt:P345 """+imdb_id+""" .
        ?movie wdt:P840 ?place.
        SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
    }"""
  
  r = requests.get(endpoint, params = {'format': 'json', 'query': query})
  if 'json' in r.headers.get('Content-Type'):
    data = r.json()
    if len(data['results']['bindings'])>0:
        return data['results']['bindings'][0]['placeLabel']['value']
    else:
        return None
  else: 
    return None
````

````python
def get_geonames_id(imdb_id):   #A function to return the geoname Id based on an imdb id
  imdb_id = '"'+imdb_id+'"'
  endpoint = "https://query.wikidata.org/bigdata/namespace/wdq/sparql"
  query = """ SELECT ?country ?countryLabel ?GeonamesID 
    WHERE {
        ?movie wdt:P345 """+imdb_id+""" .
        ?movie wdt:P840 ?country.
        ?country wdt:P1566 ?GeonamesID.
    }"""
  r = requests.get(endpoint, params = {'format': 'json', 'query': query})
  if 'json' in r.headers.get('Content-Type'):
    data = r.json()
    for entity in data["results"]["bindings"]:
      print("URI of country = ",entity["country"]["value"])
      print("Geonames id = ",entity["GeonamesID"]["value"])
      print("         ")
      GeonId = entity["GeonamesID"]["value"]
      if (GeonId is not None):
            return GeonId
      else:
            return None
  else: 
    return None
````

````python
from rdflib import Graph, Literal, URIRef, Namespace
from rdflib.namespace import XSD
import requests
from time import time,sleep

g = Graph()
ns1 = Namespace('http://example.org/ns/')
ns2 = Namespace('http://example.org/props/')

endpoint = "http://api.geonames.org/hierarchyJSON"
my_username = "aleksandrakapa"

g.parse('ex3-myOnto_new.nt',format='nt')

for a,b,c in g:
    if (str(b) == "http://www.w3.org/1999/02/22-rdf-syntax-ns#type" and str(c) == "http://example.org/ns/Movie"):
        imdb_id = str(a[22:])
        placeLabel = get_place_by_imdb_id(imdb_id)
        if (placeLabel is not None):  #If the movie has narrative location
            placeLabel = placeLabel.split(', ')
            placeLabel=[g.replace(' ','_') for g in placeLabel]
            a = URIRef(ns1+imdb_id)
            b = URIRef(ns2+'hasNarrativeLocation')
            for s in placeLabel:
              c = URIRef(ns1+s)
              print(imdb_id,"hasNarrativeLocation",s)
              g.add((a,b,c))
            #We have added the information to our knowledge
            #Now we want to get the geonames id
            global GeoId
            GeoId = get_geonames_id(imdb_id)
            if (GeoId is not None):
                
                #Now we have the geonames id
       
                r = requests.get(endpoint, params = {'geonameId': GeoId, 'username': my_username})
                data = r.json()

#                 print(len(data["geonames"]))
                print("The following is the top-to-bottom hierarchy for the Geonames entity with ID: " + GeoId)
                i=0
                for entity in data["geonames"]:
#                   print("----------------")
#                   print(entity["geonameId"])  #prints the geoname id of the top
#                   print(entity["name"]) #prints the name of the place
                      if (i!=0):
                        entity["name"]=entity["name"].replace(' ','_')
                        a = URIRef(ns1 + str(placeLabel))   #we want to add the hierarchy in the knowledge
                        b = URIRef(ns2 + 'isLocatedIn')
                        c = URIRef(ns1 + entity["name"])
                        print(entity["name"],'isLocatedIn',str(placeLabel))
                        g.add((c,b,a))
                      placeLabel = entity["name"]
                      i=i+1
                print("           ")      
             
        else:
            print("This movie hasn't got a specified narrative location")
        sleep(2)
        
````

````python
import nltk 
from bs4 import BeautifulSoup
import re
from nltk.tokenize.toktok import ToktokTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


nltk.download(["stopwords", "vader_lexicon", "wordnet", "punkt"])
````

Τα κείμενα που θέλουμε να αναλύσουμε (είτε οι περιγραφές είτε τα σχόλια των χρηστών) μπορεί να περιέχουν και πληροφορία την όποια δεν χρειαζόμαστε, όπως html tags ή χαρακτήρες όπως παρενθέσεις, brackets κ.ά., οι οποίοι δεν προσδίδουν κάποια πληροφορία. Για αυτό πριν την ταξινόμηση των κειμένων οι χαρακτήρες αυτοί πρέπει να αφαιρεθούν. 

Για την αφαίρεση των html tag χρησιμοποιούμε την βιβλιοθήκη **bs4** που περιέχει την συνάρτηση **BeautifulSoup**, ενώ για την αφαίρεση των υπόλοιπων χαρακτήρων έχουμε κατασκευάσει εμείς αντίστοιχες συναρτήσεις. 

Για παράδειγμα στο δικό μας Task δεν μας ενδιαφέρει αν τα γράμματα είναι πεζά ή κεφαλαία όποτε επιλέγουμε εμείς να γίνουν όλα πεζά. Επίσης μπορεί να μην μας ενδιαφέρει το περιεχόμενο μέσα σε παρενθέσεις ή αγκύλες όποτε και αυτό μπορούμε αν θέλουμε να το αφαιρέσουμε με την συνάρτηση `remove_between_square_brackets()`. Το ίδιο και με τους ειδικούς χαρακτήρες ή αριθμούς μέσω της συνάρτησης `remove_special_characters()`.



````python
#Removing the html strips
def strip_html(text):
    '''
    Συνάρτηση που αλλάζει μετατρέπει αφαιρεί το html περιεχόμενο μέσα από το κείμενο
    '''
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()


def lower_text(text):
    '''
    Συνάρτηση που αλλάζει μετατρέπει όλους τους χαρακτηρες σε πεζα
    '''
    return text.lower()

def remove_between_square_brackets(text):
    '''
    Συνάρτηση που αφαιρεί το περιεχόμενο που βρίσκεται εντός bracket
    '''
    return re.sub('\[[^]]*\]', '', text)

def remove_special_characters(text, remove_digits=True):
    '''
    Συνάρτηση που αφαιρεί ειδικούς χαρακτήρες ή και τα γραμματα αν θέλουμε
    '''
    pattern=r'[^a-zA-z0-9\s]'
    text=re.sub(pattern,'',text)
    return text
````

Οι παραπάνω συναρτήσεις αφορούσαν το πρώτο στάδιο της προ- επεξεργασίας του κειμένου, το οποίο είναι και το πιο απλό. Στο δεύτερο στάδιο, θα ασχοληθούμε με τα πιο ουσιαστικά βήματα της προ-επεξεργασίας. Αυτά αφορούν την αφαίρεση των λέξεων που δεν περιέχουν ιδιαίτερη πληροφορία όπως οι συνδετικές λέξεις και την κανονικοποιήση των λέξεων σε μια βασική μορφή με την αφαίρεση των καταλήξεων τους.

Για την αφαίρεση των stopwords θα χρησιμοποιήσουμε την βιβλιοθήκη nltk η οποία μας παρέχει μια πλήρη λίστα με όλες αυτές τις λέξεις για διάφορες γλώσσες. Τέλος για την κανονικοποιήση των λέξεων παρέχουμε 2 συναρτήσεις που χρησιμοποιούν την PorterStemmer και WordNetLemmatizer της ίδια βιβλιοθήκης. 

Περισσότερες πληροφορίες για την λειτουργία των μεθόδων αυτών μπορείτε να βρείτε [εδώ](https://www.tutorialspoint.com/natural_language_toolkit/natural_language_toolkit_stemming_lemmatization.htm).


````python
tokenizer=ToktokTokenizer() #για σπάσουμε το κείμενο σε tokens

stopword_list=nltk.corpus.stopwords.words('english') # η λίστα των stopwords

#removing the stopwords
def remove_stopwords(text, is_lower_case=False):
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    if is_lower_case:
        filtered_tokens = [token for token in tokens if token not in stopword_list]
    else:
        filtered_tokens = [token for token in tokens if token.lower() not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)    
    return filtered_text


def stemmer(text):
    words = nltk.word_tokenize(text)
    ps=nltk.porter.PorterStemmer()
    text= ' '.join([ps.stem(word) for word in words])
    return text


lm = WordNetLemmatizer()
def lemmatizer(text):
    words = nltk.word_tokenize(text)
    text= ' '.join([lm.lemmatize(word) for word in words])
    return text
````

Όλες οι παραπάνω συναρτήσεις δέχονται σαν είσοδο ένα text και επιστρέφουν ένα text. Έτσι με τον τρόπο αυτό μπορούμε να κατασκευάσουμε μια αλυσίδα ενεργειών που θέλουμε να εφαρμόσουμε σε ένα κείμενο. H παρακάτω συνάρτηση δέχεται σαν είσοδο ένα κείμενο και μια λίστα με συναρτήσεις και τις καλεί σειριακά δημιουργώντας ένα pipeline προ-επεξεργασίας του κειμένου.

````python
def preprocess_text(text, pipeline):
    '''
    Η συνάρτηση η οποία καλεί όλες τις μεθόδους του pipeline ώστε να γίνει
    η σωστή προεπεξεργασία του κειμένου εισόδου
    '''
    for method in pipeline:
        text = method(text)
    return text
````

````python
war_keywords = ["war", "soldier", "attack"]

war_keywords = [preprocess_text(word, pipeline) for word in war_keywords] #περνάμε και τις λέξεις από το ίδιο pipeline

def find_group(keywords, pipeline = pipeline):
    group_id = []
    for imdb_id, description in descriptions.items():
        description = preprocess_text(description, pipeline)
        for keyword in keywords:
            if keyword in description:
                group_id.append(imdb_id)
    return group_id

#We make a new function that appends a movie only if it has 2 of the keywords
def find_group2(keywords, pipeline = pipeline):
    group_id = []
    hasOne = False
    for imdb_id, description in descriptions.items():
        description = preprocess_text(description, pipeline)
        for keyword in keywords:
            if keyword in description:
                if (hasOne == True):
                    group_id.append(imdb_id)
                    hasOne = False
                else: hasOne = True
    return group_id

#We make a new function that appends a movie only if it has 3 of the keywords
def find_group3(keywords, pipeline = pipeline):
    group_id = []
    hasOne = False
    hasTwo = False
    for imdb_id, description in descriptions.items():
        description = preprocess_text(description, pipeline)
        for keyword in keywords:
            if keyword in description:
                if (hasOne == True):
                    if (hasTwo == True):
                        group_id.append(imdb_id)
                        hasOne = False
                        hasTwo = False
                    else:
                        hasTwo = True
                else: 
                    hasOne = True
    return group_id

war_movies_id = find_group2(war_keywords)
print (f"Number of war movies: {len(war_movies_id)}")
````

**3ος recommender**

Επιστρέφει ταινίες ίδιας χώρας παραγωγής με αυτές που αρέσουν στον χρήστη, περιέχει τουλάχιστον 2 ηθοποιούς που παίζουν σε ταινίες που αρέσουν στον χρήστη και έχουν θετικά σχόλια.

````python
def third_recommender(graph,user):
  ans=graph.query(
      """
      PREFIX exp: <http://example.org/props/>
      PREFIX ex: <http://example.org/ns/>
      
      SELECT DISTINCT ?movie
        WHERE { 
          ex:"""+user+""" exp:Likes ?likemovie.
          ?movie exp:hasComments "Good"^^xsd:string.
          ?movie ex:hasActor ?actora.
          ?movie ex:hasActor ?actorb.
          ?likemovie ex:hasActor ?actora.
          ?likemovie ex:hasActor ?actorb.
          FILTER (?actora != ?actorb)
        }""")
  
  ans = [a['movie'].toPython().replace(ns1,'') for a in ans]
  return ans
````
